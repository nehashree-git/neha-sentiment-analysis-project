# Outlier Detection (Ratings)

```python
Q1 = df['rating'].quantile(0.25)
Q3 = df['rating'].quantile(0.75)
IQR = Q3 - Q1
lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR

outliers = df[(df['rating'] < lower) | (df['rating'] > upper)]
print("Outliers in rating:", len(outliers))
```
